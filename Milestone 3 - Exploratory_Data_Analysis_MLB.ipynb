```python
from google.cloud import storage
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import glob
import os
```

```python
client = storage.Client()
bucket_name = 'my-project-bucket-jl'
bucket = client.get_bucket(bucket_name)
elements = bucket.list_blobs()
csv_files=[a.name for a in elements]
```

```python
csv_files
```

```python
mlb_gameday_list = []
for file in csv_files:
    try:
        filename = 'gs://my-project-bucket-jl/' + file + ''
        df = pd.read_csv(filename)
        if df.empty:
            continue
        else:
            filename = 'gs://my-project-bucket-jl/' + file + ''
            print(filename)
            df = pd.read_csv(filename)
            mlb_gameday_list.append(df)
    except pd.errors.EmptyDataError:
        continue
```

```python
merged_mlb_gameday_list_df = pd.concat(mlb_gameday_list,ignore_index = True)
```

```python
#function to get string and object statistics including unique values, most frequent value, by what 
#frequency, number of missing values and length of column
def object_statistics(df):
    results = {}
    for col in df.select_dtypes(include='object'):
        results[col] = {'unique_values': df[col].nunique(),
                        'most_frequent': df[col].value_counts().index[0],
                        'frequency': df[col].value_counts().iloc[0],
                        'missing_values': df[col].isnull().sum(),
                        'lenght': len(df[col])
        }

    return results
```

```python
object_statistics(merged_mlb_gameday_list_df)
```
