```python
#list of csv files in 'my-project-bucket-jl' bucket
client = storage.Client()
bucket_name = 'my-project-bucket-jl'
bucket = client.get_bucket(bucket_name)
elements = bucket.list_blobs()
csv_files=[a.name for a in elements]
```

```python
#file path for each csv file 
mlb_gameday_list = []
for file in csv_files:
    try:
        filename = 'gs://my-project-bucket-jl/' + file + ''
        df = pd.read_csv(filename)
        if df.empty:
            continue
        else:
            filename = 'gs://my-project-bucket-jl/' + file + ''
            print(filename)
            df = pd.read_csv(filename)
            mlb_gameday_list.append(df)
    except pd.errors.EmptyDataError:
        continue
```

```python
#cleaning data
for file in csv_files:
    try:
        filename = 'gs://my-project-bucket-jl/' + file + ''
        df = pd.read_csv(filename)
        if df.empty:
            continue
        else:
            filename = 'gs://my-project-bucket-jl/' + file + ''
            print(filename)
            mlb_gameday_df = pd.read_csv(filename)
            mlb_gameday_new_df = mlb_gameday_df.drop(['game_date','release_speed','release_pos_x','release_pos_z',
                                                      'player_name','description','spin_dir','spin_rate_deprecated',
                                                      'break_angle_deprecated','break_length_deprecated','des',
                                                      'game_type','home_team','away_team','hit_location','bb_type',
                                                      'pfx_x','pfx_z','plate_x','plate_z','hc_x','hc_y',
                                                      'tfs_deprecated','tfs_zulu_deprecated','fielder_2','umpire',
                                                      'sv_id','hit_distance_sc','launch_speed','launch_angle',
                                                      'effective_speed','release_spin_rate','release_extension',
                                                      'pitcher.1','fielder_2.1','fielder_3','fielder_4','fielder_5',
                                                      'fielder_6','fielder_7','fielder_8','fielder_9',
                                                      'release_pos_y','estimated_ba_using_speedangle',
                                                      'estimated_woba_using_speedangle', 'woba_value','woba_denom',  
                                                      'babip_value', 'iso_value','launch_speed_angle', 'pitch_name',
                                                      'bat_score','fld_score','post_bat_score','post_fld_score',
                                                      'if_fielding_alignment','of_fielding_alignment','spin_axis',
                                                      'delta_home_win_exp','delta_run_exp','bat_speed',
                                                      'swing_length'],axis=1)
            mlb_gameday_new_df.dropna(subset=['pitch_type'], inplace=True)
            storage_client = storage.Client()
            bucket = storage_client.get_bucket('my-project-bucket-jl')
            filedata = mlb_gameday_new_df.to_parquet(index=0)
            blob = bucket.blob(f"revised_{mlb_gameday}.parquet")
            blob.upload_from_filename(filedata, content_type = 'application/octet=stream')
    except pd.errors.EmptyDataError:
        continue
```
